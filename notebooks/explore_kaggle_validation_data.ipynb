```python
import json
from itertools import chain

import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
validation_labels = pd.read_csv("../data/kaggle/validation.csv")
validation_labels.describe()
```

```python
validation_labels["Usage"].value_counts()
```

```python
label_strings = list(
    chain(
        *list(
            map(
                lambda s: list(map(lambda _s: _s.strip(), s.split("|"))),
                validation_labels["PredictionString"].values,
            )
        )
    )
)
label_strings
```

```python
labels, counts = np.unique(label_strings, return_counts=True)
```

```python
top_idxs = np.argsort(-counts)
labels[top_idxs]
```

```python
ids = validation_labels["Id"].values
```

```python
data = {
    "ids": [],
    "text": [],
    "labels": [],
}

for id in tqdm(ids):
    with open(f"../data/kaggle/validation/{id}.json") as fp:
        data["text"].append(json.load(fp))
        data["ids"].append(id)
        data["labels"].append(
            validation_labels[validation_labels["Id"] == id]["PredictionString"].values[
                0
            ]
        )
```

```python
df = pd.DataFrame(data)
df.to_pickle("../data/kaggle/validation_dataframe.pkl")
```
